```python
import tensorflow as tf
import pandas as pd 
import numpy as np
import seaborn as sns

```

```python
df = pd.read_csv("stats_females.csv")
df.head()
```

```python
sns.pairplot(df)
```

## Select your features and labels

```python
from sklearn.model_selection import train_test_split

X = df[['momheight','dadheight']].values
y = df['Height'].values
```

```python
X
X.shape
```

```python
y
```

## Convert your data into Training and Testing

```python
X_train ,X_test, y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X.shape
```

## Normalize the Data before processing

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Model Creation 

```python
model = Sequential()
model.add(Dense(4,activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1))
model.compile(optimizer = 'rmsprop',loss='mse')
model.fit(x = X_train, y = y_train,epochs=500)
```

```python
losshistory = pd.DataFrame(model.history.history)
```

```python
losshistory.plot()
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
test_predictions = pd.Series(test_predictions.reshape(65,))

pred_df = pd.DataFrame(y_test,columns = ['y_test'])
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns = ['y_test','predictions']
```

```python
pred_df
```

```python
sns.scatterplot(x="y_test",y="predictions",data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['y_test'],pred_df['predictions'])
```

```python
mean_squared_error(pred_df['y_test'],pred_df['predictions'])
```

```python
mean_squared_error(pred_df['y_test'],pred_df['predictions'])**0.5
```
